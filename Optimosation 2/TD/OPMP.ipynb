```python
from platypus import NSGAII,Problem,Integer, unique,nondominated

def belegundu(vars):
    x1 = vars[0]
    x2 = vars[1]
    x3 = vars[2]
    x4 = vars[3] 
    x5 = vars[4] 
    x6 = vars[5] 
    x7 = vars[6] 
    x8 = vars[7]
    return [11*x1 + 4*x2 + 5*x3+x4+7*x5+13*x6+4*x7+3*x8 , 9*x1 + 8*x2 + 2*x3+2*x4+16*x5+5*x6+9*x7+4*x8],[4*x1+3*x2+2*x3+x4+8*x5+7*x6+6*x7+5*x8]

problem = Problem(8,2,1)

problem.types[:] = [Integer(0,1),Integer(0,1),Integer(0,1),Integer(0,1),Integer(0,1),Integer(0,1),Integer(0,1),Integer(0,1)]

problem.constraints[:] ="<=20"

problem.function = belegundu

problem.directions[0] = Problem.MAXIMIZE

problem.directions[1] = Problem.MAXIMIZE

algorithm = NSGAII(problem)

algorithm.run(100000)

for solution in algorithm.result:
    print(solution.objectives)
for solution in algorithm.result:
    print(solution.variables)
for solution in unique(nondominated(algorithm.result)):
    print(solution.objectives)
import matplotlib.pyplot as plt
plt.scatter([s.objectives[0] for s in algorithm.result ],[s.objectives[1] for s in algorithm.result])
plt.xlabel("$f_1(x)$")
plt.ylabel("$f_2(x)$")
plt.show()
```
