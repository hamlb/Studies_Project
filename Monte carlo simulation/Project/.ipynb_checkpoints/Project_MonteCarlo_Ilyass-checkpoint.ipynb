## Evaluation of the opportunity to enter into a new project

```python
#Package import 
import numpy as np
import matplotlib.pyplot as plt 
import scipy.stats as sps
```

```python
#Parameters
t0 = 0
t1 = 1.
t2 = 6.

X_0 = 104.6
sigma = 0.2
c = 7.67
a = 2.36
b = 4.15
q = 10
time_grid = np.array([t0, t1, t2])

M = 3
N = 2
W_1 = np.zeros(M)
W_2 = np.zeros(N)
X_1 = np.zeros(M)
X_2 = np.zeros(N)

B= np.random.beta(a, b, size=None)
```

```python
W_1 = t1 - t0

# M i.i.d. samples of X_1
G = np.random.normal(loc=0, scale=1, size=M)
    
X_1 = X_0 * np.exp(sigma * W_1*G - sigma**2*(t1)/2)
    
# To Do: simulate M*N i.i.d. samples 
# of the Brownian increment Delta W
Delta_W = np.sqrt(t2 - t1) * np.random.normal(loc=0, scale=1, size=(M,N))
    
for m in range(M):
    # We work with one-dimensional arrays of size N
    exponential_factor = np.exp(sigma * Delta_W[m] - 0.5*sigma*sigma*(t2 - t1))
    X_2 = X_1[m] * exponential_factor
```

```python
print(X_1)
print(X_2)
```

```python
# Defining Q = q(1+B);  q(t)
B= np.random.beta(a, b, size=(M,N))
Q = q*(1+B)
```

```python
# Defining U_1(X_1) = E[f_2(Q,X_2)|X_1]
integral = 0.5 * c * (Q + q) * (t2 - t1) 
f_2 = Q * X_2  - integral
U_1 = np.mean(f_2, axis=1)

#plotting U_1
plt.hist(U_1)
plt.title('Values of the expected payoff U_1', fontsize='10')
plt.show()
```

```python
# The positive part function
def positive_part(x):
    return np.maximum(x,0)
```

```python
# Defining Gain_1 = E[(ùë¢_1 (ùëã_1 )‚àíùëì_1 (ùëã_1 ))+ ]
f_1 = q * X_1 - c * q * t1
Gain_X1 = positive_part(U_1 - f_1)

# Plotting Gain_X1
plt.figure(figsize=(5,5))
plt.hist(Gain_X1)
plt.xlabel('Gain exceedance')
plt.ylabel('Frequency')

plt.title('Distribution of the gain exceedance at time t1', fontsize='12')

plt.show()

#print(gain_exceedance)
```

```python
# Defining the positive set E1 = {x : u1(x) ‚àí f1(x) ‚â• 0}
positive_set = []
for i in range(M):
    if U_1[i] - f_1[i]>=0:
        positive_set.append(X_1[i])

Gain_X1 = U_1 - f_1

# Plotting positive set points
plt.figure(figsize=(10,5))
plt.scatter(X_1,Gain_X1, marker='.')
plt.plot(X_1, np.zeros(M), color="red" , label="f1")
plt.title('Positive Set Points', fontsize='12')
plt.xlabel('X1')
plt.ylabel('Gain')

plt.show()
print("The positive set is given by the following array: ")
print(X_1)
```
